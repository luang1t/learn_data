```python
# 1. Importa a biblioteca para baixar datasets do Kaggle
import kagglehub

# 1.2 Baixar (ou atualiza) o dataset que queremos trabalhar localmente
path = kagglehub.dataset_download("die9origephit/fifa-world-cup-2022-complete-dataset")

# 1.3 Aqui conseguimos ver o caminho do nosso dataset, de onde iremos extrair
print("caminho para o dataset que iremos trabalha:\n",path)
```

```python
# 2. Biblioteca para interação com o sistema operacional e arquivos
import os

# 2.1 os.listdir() = Lista todos os arquivos presentes no diretorio especificado
arquivos = os.listdir(path)

print(arquivos)
```

```python
# 3. Biblioteca para manipulação e análise de dados
import pandas as pd

# 3.1 Define o caminho completo para o arquivo CSV das partidas
arquivo_final = os.path.join(path, 'Fifa_world_cup_matches.csv')

# 3.2 Carrega os dados do arquivo CSV para um DataFrame do Pandas
df = pd.read_csv(arquivo_final)
```

```python
# 4. Visualiza (por padrão) as 5 primeiras linhas para garantir a integridade do arquivo
df.head()
```

```python
# 5. Visualiza (por padrão) as 5 últimas linhas para garantir a integridade do final do arquivo
df.tail()
```

```python
# 6. Verifica a estrutura do dataset: tipos de dados (Dtypes) e presença de valores nulos (NaN)
df.info()
```
