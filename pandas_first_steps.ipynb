```python
# 1. Importa a biblioteca para baixar datasets do Kaggle
from kagglehub import dataset_download

# 1.2 Baixar (ou atualiza) o dataset que queremos trabalhar localmente
path = dataset_download("die9origephit/fifa-world-cup-2022-complete-dataset")

# 1.3 Aqui conseguimos ver o caminho do nosso dataset, de onde iremos extrair
print("caminho para o dataset que iremos trabalha:\n",path)
```

```python
# 2. Biblioteca para interação com o sistema operacional e arquivos
import os

# 2.1 os.listdir() = Lista todos os arquivos presentes no diretorio especificado
arquivos = os.listdir(path)

print(arquivos)
```

```python
# 3. Biblioteca para manipulação e análise de dados
import pandas as pd

# 3.1 Define o caminho completo para o arquivo CSV das partidas
arquivo_final = os.path.join(path, 'Fifa_world_cup_matches.csv')

# 3.2 Carrega os dados do arquivo CSV para um DataFrame do Pandas
df = pd.read_csv(arquivo_final)
```

Análise da Estrutura (info):

Dimensões: Temos 64 linhas (jogos) e 88 colunas.

Tipos de Dados:

int64: Dados numéricos (gols, posse de bola, etc).

object: Texto ou misto. Atenção: A coluna date está como object (texto), mas idealmente deveria ser convertida para datetime para conseguirmos filtrar por tempo.

Qualidade: A coluna Non-Null Count nos dirá se temos buracos na base que precisam ser tratados.

Limpeza -> Exploração -> Visualização -> INSIGHTS

```python
# 1. Verifica se existe dados faltando (Células vazias).
print(df.isnull().sum())
```

```python
# 2. Verifica duplicatas (Partidas repetidas)
print(df.duplicated().sum())
```

```python
# 3. Correção da data (não está como número mas sim como string)

# 3.1 Dicionário para mapear o Mês em inglês eliminando o locale problem
meses_maps = {
    'JAN':'01', 'FEV':'02','MAR':'03','APR':'04','MAY':'05','JUN':'06',
    'JUL':'07','AUG':'08','SEP':'09','OCT':'10','NOV':'11','DEC':'12'
}

# 3.2 Fazendo a substituição na coluna de texto
# OBS: Regex=True garante que ele encontre o texto mesmo se houver espaços estranhos
for mes_txt, mes_num in meses_maps.items():
  df['date'] = df['date'].str.replace(mes_txt,mes_num, regex=True)

#3.3 Data atualizada no formato "26 12 2000" (apenas números)
# OBS: Agora é possível converter especificando o formato exato: Dia Mês Ano
df['date'] = pd.to_datetime(df['date'], format='%d %m %Y')
```

```python
print(df.info())
```

```python
# 4. Correção da posse de bola
value_for_change = ['possession team1','possession team2', 'possession in contest']
df[value_for_change] = df[value_for_change].replace('%','', regex = True)
df[value_for_change] = df[value_for_change].astype(float)
```
